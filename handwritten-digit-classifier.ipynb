# Handwritten Digit Classifier

```python
import tensorflow as tf
print("TensorFlow works!")
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
print("Got 60,000 pictures of numbers!")
print(f"First number is: {y_train[0]}")
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0], cmap='gray')
plt.title(f"Number: {y_train[0]}")
plt.show()
```

```python
# Build + Train model (copy ALL this into NEW cell)
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=3)  # 3 minutes training
```

```python
# Test on unseen data
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_acc:.2%}")
print(f"Model Performance: {test_acc:.2%} on handwritten digit recognition")
```
